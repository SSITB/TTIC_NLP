# Section 1.4 Error Analysis

### Multiple Sentiments/Most important sentiment at end.

- Sentence: "With the exception of some fleetingly amusing improvisations by Cedric the Entertainer as Perry 's boss,there is n't a redeeming moment here."
    - Model Classification: 2
    - Gold Standard: 0
    - Reason: - "isn't a redeeming moment" turns the sentence into a criticism. Previous words would be considered as praise as thus have positive weights. The gold standard annotation is valid.
    

- Sentence: "There 's enough melodrama in this Magnolia Primavera to make PTA proud yet director Muccino 's characters are less worthy of Puccini than they are of "'daytime television.'
    - Model Classification: 2
    - Gold Standard: 0
    - Reason: Sentiment at end of sentence & multiple sentiments.Worthy and proud have high positive weights.


- Sentence: 'A hamfisted romantic comedy that makes our girl the hapless facilitator of an extended cheap shot across the Mason-Dixon line.'
    - Model Classification: 2
    - Gold Standard: 0
    - Reason: negative sentiment towards the end. Comedy and Romantic have high positive weights. Hapless not in training set.


- Sentence: "Impostor has a handful of thrilling moments and a couple of good performances, but the movie does n't quite fly."
    - Model Classification: 2
    - Gold Standard: 0
    - Reason: Negative sentiment appears at the end.


- Sentence: 'A difficult, absorbing film that manages to convey more substance despite its repetitions and inconsistencies than do most films than are far more ''pointed and clear.'
    - Model Classification: 0
    - Gold Standard: 2
    - Reason: This is a very complex sentence with words that have a primarily negative sentiment, however, their order implies high praise.
    
    
- Sentence: 'The only excitement comes when the credits finally roll and you get to leave the theater.'
    - Model Classification: 2
    - Gold Standard: 0
    - Reason: Negative sentiment used at the end of sentence. "Excitement" has positive weight but is used humorously.


- Sentence: "Do n't be fooled by the impressive cast list - Eye See You is pure junk."
    - Model Classification: 2
    - Gold Standard: 0
    - Reason: negation using "junk" at the very end, sentence has high positive weights for "impressive" and "pure"

### Form of negation
- Sentence: "There 's a solid woman - finding-herself story somewhere in here , but you 'd have to dig pretty deep to uncover it."
    - Model Classification: 2
    - Gold Standard: 1
    - Reason: use of the word solid is what tripped the classifier into assigning label 2. Solid is only associated with positive labels in training. Use of the word dig and pretty (positive weight) are used in the ngative sense. As such, the annotation label is also incorrect.
    
### Incorrect Annotation

- Sentence: "There 's just no currency in deriding James Bond for being a clichéd , doddering , misogynistic boy 's club."
    - Model Classification: 0
    - Gold Standard: 1
    - Reason: Incorrect annotation. Model performed superb in this case!


- Sentence: 'The ga-zillionth airhead movie about a wife in distress who resorts to desperate measures.'
    - Model Classification: 0
    - Gold Standard: 1
    - Reason: This is a negative comment, not a neutral comment. Model did a good job of picking that up!


- Sentence: 'A gripping , searing portrait of a lost soul trying to find her way through life.'
    - Model Classification: 2
    - Gold Standard: 1
    - Reaosn: Very impressive result! Training data has gripping labelled as a 2. This is not a neutral sentence, but rather, high praise!


- Sentence: "It 's mighty tedious for the viewer who has to contend with unpleasant characters , hit-and-miss performances and awkwardly staged scenes."
    - Model Classification: 0
    - Gold Standard: 1
    - Reason: This is clearly a sentence with negative sentence and not a neutral sentence. Model performed well here.


- Sentence: "Let 's hope -- shall we ?"
    - Model Classification: 2
    - Gold Standard: 1
    - Reason: To hope is not neutral, it is inherently positive.


- Sentence: 'If this story must be told and retold -- and indeed it must -- then The Grey Zone is to be lauded for finding a new and ingenious angle.'
    - Model Classification: 2
    - Gold Standard: 1
    - Reason: **AMAZING MODEL RESULT** - this sentence is clearly positive and not neutral.

### Human Intuition Required
- Sentence: "To say this was done better in Wilder 's Some Like It Hot is like saying the sun rises in the east."
    - Model Classification: 2
    - Gold Standard: 0
    - Reason: Requires external knowledge to know this is a negative connotation. The sun rises in the east, and so this movie is contributing nothing new. This is a case of complicated use of launguage and sarcasm/humor.


- Sentence: 'This remake gets all there is to get out of a peculiar premise with promise : Al Pacino loathing Robin Williams.'
    - Model Classification: 0
    - Gold Standard: 1
    - Reason: In this context, loathing was meant to describe the central "emotion" of the movie that was portrayed very well. As such, this sentence could be either positive or negative.


- Sentence: 'Scorsese at his best makes gangster films that are equally lovely but also relentlessly brutal and brutally intelligent ; Perdition , meanwhile , ''reads more like Driving Miss Daisy than GoodFellas.'
    - Model Classification: 2
    - Gold Standard: 1
    - Reason: Firstly, this sentence has incorrect annotation. It should be 0. Additionally, human intuition si required to know the "driving miss daisy" is diametrically opposed to "GoodFellas".


### Word Sense Ambiguity

- Sentence: 'Skins has a right to yawp , and we have a right to our grains of salt.'
    - Model Classification: 2
    - Gold Standard: 1
    - Reason: Model cannot discern th neutral use of the word right - as in "obligation" as opposed to "correct"


- Sentence: 'While the Resident Evil games may have set new standards for thrills , suspense , and gore for video games , the movie really only succeeds in the ''third of these.'
    - Model Classification: 2
    - Gold Standard: 0
    - Reason: Very complex sentence, use of the word "succeeds" to give negative sentiment.


- Sentence: "It 's the perfect kind of film to see when you do n't want to use your brain."
    - Model Classification: 0
    - Gold Standard: 1
    - Reason: The sentence is very neutral in that you cannot infer if the reviewer meant "dont use your brain" as either relaxing or horrendous. In this case, use of the word "perfect" is to describe a balance between sentiments, rather than a strictly positive sentiment. This is a difficult case. "do n't" is present in either class 0 or class 2, never in class 1.
