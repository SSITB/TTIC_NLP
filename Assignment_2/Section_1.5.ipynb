I have created the following feature template:

$f^{u,b}(x,y) = \mathbb{I}[y = \operatorname{label}]\wedge \mathbb{I}[x\ \operatorname{contains}\ (word_1, word_2)]$

The above is a bigram template.

For example, the sentence:

"The Rock is destined to be the 21st Century 's new \`\` Conan '' and that he 's going to make a splash even greater than Arnold Schwarzenegger , Jean-Claud Van Damme or Steven Segal."

Produces the following bigrams:

('The', 'Rock'),
 ('Rock', 'is'),
 ('is', 'destined'),
 ('destined', 'to'),
 ('to', 'be'),
 ('be', 'the'),
 ('the', '21st'),
 ('21st', 'Century'),
 ('Century', "'s"),
 ("'s", 'new'),
 ('new', '\`\`'),
 ('\`\`', 'Conan'),
 ('Conan', "''"),
 ("''", 'and'),
 ('and', 'that'),
 ('that', 'he'),
 ('he', "'s"),
 ("'s", 'going'),
 ('going', 'to'),
 ('to', 'make'),
 ('make', 'a'),
 ('a', 'splash'),
 ('splash', 'even'),
 ('even', 'greater'),
 ('greater', 'than'),
 ('than', 'Arnold'),
 ('Arnold', 'Schwarzenegger'),
 ('Schwarzenegger', ','),
 (',', 'Jean-Claud'),
 ('Jean-Claud', 'Van'),
 ('Van', 'Damme'),
 ('Damme', 'or'),
 ('or', 'Steven'),
 ('Steven', 'Segal'),
 ('Segal', '.')\
 
 I still use a feature count cutoff of 1.
 
 This procedure creartes 132,538 features from the training data.
